```python
import pandas

##this imports your graphing tool, seaborn as 'sns'
import seaborn as sns
sns.set_style('darkgrid')

##this assigns the filename we're trying to load in to a string variable
csvfile = 'Travel Pony Facebook.csv'

###convert to dataframe 
df = pandas.read_csv(csvfile)


#histogram of Amount Spent using Seaborn
hist = sns.distplot(df['Amount Spent (USD)'])

#multiple regression, outcome == Amount Spent
#predictors == Campaign Name,Reach,Frequency,Impressions,Clicks,
#Unique Clicks, Page Likes, Page Engagement, Post Engagement
#Post Likes, Post Comments, Post Shares, Photo Views
#Website Clicks
import numpy
import statsmodels.api as sm
```

```python
#code Campaign Name to category
df['Campaign Name']=df['Campaign Name'].astype('category')
df['Campaign Name']=df['Campaign Name'].cat.codes

X = df[["Campaign Name","Reach","Frequency","Impressions","Clicks","Unique Clicks", "Page Likes", "Page Engagement", "Post Engagement","Post Likes", "Post Comments", "Post Shares", "Photo Views"]]
y = df["Amount Spent (USD)"]
```

```python
#create the model 
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```

```python
#look at coefficients for each predictor variables 
#rank the three highest predictive values
linreg = sm.OLS(y, X).fit()
predictions = linreg.predict(X)
linreg.summary()
```
