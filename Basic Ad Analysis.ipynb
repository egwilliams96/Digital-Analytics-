```python
import pandas

##this imports your graphing tool, seaborn as 'sns'
import seaborn as sns

##this assigns the filename we're trying to load in to a string variable
csvfile = 'Travel Pony Facebook.csv'
```

```python
###convert to dataframe 
df = pandas.read_csv(csvfile)
```

```python
###First, create the analytics 'cost per impression'
### by dividing Amount Spent / Impressions. Do this 
##for each ad in your data, and save that as a column.
costperimpression = (df["Amount Spent (USD)"])/ (df["Impressions"])
df['Cost per Impression'] = (df["Amount Spent (USD)"])/ (df["Impressions"])
df.head()
```

```python
#When considering 'cost per impression', what day 
#of the week works best? (What day is it cheapest 
#to generate impressions) What day works worst? 
#(What day is most expensive?
import arrow

dt = df['Start Date']


weekdays= []
for i, row in df.iterrows():
    dt = arrow.get(row['Start Date'], 'M/D/YY')
    weekdays.append(dt.weekday())
df['weekday']= weekdays


df.groupby(['Cost per Impression','weekday'])['weekday'].count()
     
```
